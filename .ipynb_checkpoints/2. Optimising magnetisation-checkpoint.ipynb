## 2. Optimising magnetisation

This jupyter notebook is used to computationally calculate the magnetisation of the magnet we used in the experiments.

The magnetisation was found Experimentally to be 1.06 +/- 0.07 e6.

A map of B-field measurements were taken above a single magnet, (see OneDrive/Readings & Results/Magnet Map Readings folder). These B-field measurements have 2 axis of symmetry (along the x-axis and y-axis, see diagram on "Magnet Map Results" excel doc) meaning that, for every B-field measurement that isn't on the x-axis or y-axis, there are 3 other measurements of the same B-field.

Therefore, every B-field measurement was averaged with any other measurement that should give the same reading. This created a new b-field map which is saved in data/magnet_reading_mean.csv. 

Note that every B-field measuresument is only a measure of the vertical component of the B-field at that point.

This Jupyter notebook computationally calculates the magnetisation of the magnet using these b-field readings.

```python
import magnet_simulator as magsim
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# load in data
readings = pd.read_csv('data/magnet_reading_mean.csv')

# format the x, y, z columns in metres
readings['x'] = readings['x'] * 1e-3
readings['y'] = readings['y'] * 1e-3
readings['z'] = readings['z'] * 1e-3

readings.head()
```

```python
# this DataFrame will hold the results
results = pd.DataFrame(columns=['magnetisation (*1e6)', 'SSR'])

magnetisations = np.array([]) * 1e6  # the magnetisations to test

for magnetisation in magnetisations:
    mag = magsim.Holed10mmMagnet(
        magnetisation=[0, 0, magnetisation], 
        cubic_section_sizes=0.4e-3  # optimum section size as found from 'Optimising magnet segment size'
    )  # create a cube magnet of size 1cm, with north pole facing up
    
    B_predicted = np.array([])  # stores the predicted B-field, each element refers to the corresponding position in
    # the results DataFrame
    
    # loop over every measurement in the results DataFrame and calculate the B-field at that point
    for index, row in readings.iterrows():
        B_predicted = np.append(
            B_predicted, 
            mag.B([row['x'], row['y'], row['z']])[2]
        )
        
    results = pd.concat([
        results,
        pd.DataFrame({
            'magnetisation (*1e6)': [magnetisation * 1e-6],
            'SSR': [np.sum((readings['B_mean'] - B_predicted)**2)]
        })
    ], ignore_index=True)
    
# read in the already saved results, sort the data, and export it back out
results = pd.concat([results, pd.read_csv('data/2.1 optimising magnetisation - finding section size.csv')]).sort_values(by='magnetisation (*1e6)')
# results = results.drop_duplicates(subset=['magnetisation (*1e6)']).reset_index(drop=True)
# results.to_csv('data/2.1 optimising magnetisation - finding section size.csv', index=False)

results
```

magnetisation is 1.107e6

```python
plt.figure(figsize=(7,3))
plt.scatter(results["magnetisation (*1e6)"],results["SSR"])

plt.show()
```
